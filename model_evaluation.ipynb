At Gives You Paws, anyone can upload pictures of their cats or dogs. The photos are then put through a two step process. First an automated algorithm tags pictures as either a cat or a dog (Phase I). Next, the photos that have been initially identified are put through another round of review, possibly with some human oversight, before being presented to the users (Phase II).

Several models have already been developed with the data, and you can find their results here.

Given this dataset, use pandas to create a baseline model (i.e. a model that just predicts the most common class) and answer the following questions:

In terms of accuracy, how do the various models compare to the baseline model? Are any of the models better than the baseline?

```python
paws_df = pd.read_csv('gives_you_paws.csv')
paws_df.info()
paws_df.head()
```

```python
vs_model_one = pd.crosstab(paws_df.actual, paws_df.model1)
vs_model_two = pd.crosstab(paws_df.actual, paws_df.model2)
vs_model_three = pd.crosstab(paws_df.actual, paws_df.model3)
vs_model_four = pd.crosstab(paws_df.actual, paws_df.model4)
```

```python
paws_df['baseline'] = 'dog'
paws_df.head()
```

```python
paws_avb_acc = (paws_df.actual == paws_df.baseline).mean()
paws_av2_acc = (paws_df.actual == paws_df.model2).mean()
paws_av3_acc = (paws_df.actual == paws_df.model3).mean()
paws_av4_acc = (paws_df.actual == paws_df.model4).mean()

print(f'Paws Baseline accuracy: {paws_avb_acc:.2%}')
print(f'Paws Model 1 accuracy: {paws_av2_acc:.2%}')
print(f'Paws Model 2 accuracy: {paws_av2_acc:.2%}')
print(f'Paws Model 3 accuracy: {paws_av3_acc:.2%}')
print(f'Paws Model 4 accuracy: {paws_av4_acc:.2%}')
print('---------------------------------')

paws_sub1 = paws_df[paws_df.model1 == 'dog']
model1_prcsn = (paws_sub1.actual == paws_sub1.model1).mean()
paws_sub2 = paws_df[paws_df.model2 == 'dog']
model2_prcsn = (paws_sub2.actual == paws_sub2.model2).mean()
paws_sub3 = paws_df[paws_df.model3 == 'dog']
model3_prcsn = (paws_sub3.actual == paws_sub3.model3).mean()
paws_sub4 = paws_df[paws_df.model4 == 'dog']
model4_prcsn = (paws_sub4.actual == paws_sub4.model4).mean()
paws_sub_base = paws_df[paws_df.baseline == 'dog']
baseline_prcsn = (paws_sub_base.actual == paws_sub_base.baseline).mean() 

#Baseline precision is redundant here, but, in general, calculate it anyway, just in case baseline happens to be the negative case


print(f'Model 1 precision: {model1_prcsn:.2%}')
print(f'Model 2 precision: {model2_prcsn:.2%}')
print(f'Model 3 precision: {model3_prcsn:.2%}')
print(f'Model 4 precision: {model4_prcsn:.2%}')
print(f'Baseline precision: {baseline_prcsn:.2%}')
print('---------------------------------')

paws_rec = paws_df[paws_df.actual == 'dog']
model1_rec = (paws_rec.actual == paws_rec.model1).mean()
model2_rec = (paws_rec.actual == paws_rec.model2).mean()
model3_rec = (paws_rec.actual == paws_rec.model3).mean()
model4_rec = (paws_rec.actual == paws_rec.model4).mean()
baseline_rec = (paws_rec.actual == paws_rec.baseline).mean()

print(f'Model 1 recall: {model1_rec:.2%}')
print(f'Model 2 recall: {model2_rec:.2%}')
print(f'Model 3 recall: {model3_rec:.2%}')
print(f'Model 4 recall: {model4_rec:.2%}')
print(f'Baseline recall: {baseline_rec:.2%}')
```
